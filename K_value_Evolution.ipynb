```python
import pandas as pd 
from Dataset import Dataset
from serverD import groundTruth_TopKItemsLiked, groundTruth_TopKItemsLiked_Cosine

dataset = Dataset("ml-100k")
trainMatrix, _, _, _ = dataset.trainMatrix, dataset.trainNegatives, dataset.testRatings, dataset.testNegatives
```

```python
K_values = range(1,200,5)
num_users = 942  
results = {}
for k in K_values:
    average_distance_users = 0
    communities = groundTruth_TopKItemsLiked(num_users, topK=k, ids_or_distances="distances")
    average_distance_per_user_list = [sum(communities[user]) / k for user in range(num_users)]
    average_distance_users +=  sum(average_distance_per_user_list) / num_users
    results[k] = average_distance_users 

results
```

```python
data = {'K values':results.keys(),'Average Similarity':results.values()}
# df =pd.DataFrame("K values","Average Similarity per user per top K"])
df = pd.DataFrame.from_dict(data)
df.plot(x = 'K values', y = 'Average Similarity', legend="MovieLens-100K")
```
